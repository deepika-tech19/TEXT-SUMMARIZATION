```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
text = df['descr'].iloc[0]

parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 3)

print("ORIGINAL TEXT:\n")
print(text)

print("\nSUMMARY:\n")
for sentence in summary:
    print(sentence)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

text = df['descr'].iloc[0]

parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

print("SUMMARY:\n")

for sentence in summarizer(parser.document, 3):
    print(sentence)
```

```python
print("Original Length:", len(text.split()))

summary_text = " ".join([str(s) for s in summarizer(parser.document, 3)])

print("Summary Length:", len(summary_text.split()))
print("\nSummary:\n", summary_text)
```

```python
import matplotlib.pyplot as plt

original_length = len(text.split())

summary_text = " ".join([str(s) for s in summarizer(parser.document, 3)])
summary_length = len(summary_text.split())

plt.figure(figsize=(6,4))
plt.bar(['Original Text', 'Summary'], [original_length, summary_length])
plt.title('Original vs Summary Length')
plt.ylabel('Number of Words')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Original Text')
plt.show()
```

```python
wordcloud = WordCloud(width=800, height=400).generate(summary_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Summary')
plt.show()
```
